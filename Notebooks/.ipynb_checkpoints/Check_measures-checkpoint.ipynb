```python
#import libraries that we need

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle

from sklearn import linear_model
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.metrics import r2_score

from sklearn.linear_model import Lasso

%matplotlib inline 
```

```python
#Read data from a csv file into a Pandas.Dataframe

df = pd.read_csv('../data/promise/ckset.csv')
df = df.drop(['name', 'version', 'namep'], axis=1)

df2 = df[df.bug == 0]

print (len(df))
```

```python
df.head()
```

```python
features = list(df)

print (features)
```

```python
l = []
ll = []
```

```python
for xstr in features:
    for ystr in features:
        
        if xstr == ystr:
            continue
        
        x = np.array(df[xstr])
        y = np.array(df[ystr])
        
        datax_train, datax_test, datay_train, datay_test = train_test_split(x, y, test_size=0.20, random_state=42)
        
        # Create a LinearRegression Model
        linear_reg = linear_model.LinearRegression()
        
        # Make data ready to fit the model
        datax_train = np.atleast_2d(datax_train).T
        datay_train = np.atleast_2d(datay_train).T
        datax_test = np.atleast_2d(datax_test).T
        datay_test = np.atleast_2d(datay_test).T
        
        # Fit the model with training data
        linear_reg.fit(datax_train, datay_train)
        
        # Get the prediction for testing data
        predict = linear_reg.predict(datax_test)
        
        # Get the score
        l.append((xstr, ystr, r2_score(datay_test, predict)))
        
        print ()
      
        plt.scatter(datax_test, datay_test,  color='black')
        plt.plot(datax_test, predict, color='blue', linewidth=3)

        #plt.xticks(())
        #plt.yticks(())

        plt.show()
```

```python
for xstr in features:
    for ystr in features:
        
        if xstr == ystr:
            continue
        
        x = np.array(df2[xstr])
        y = np.array(df2[ystr])
        
        datax_train, datax_test, datay_train, datay_test = train_test_split(x, y, test_size=0.20, random_state=42)
        
        # Create a LinearRegression Model
        linear_reg = linear_model.LinearRegression()
        
        # Make data ready to fit the model
        datax_train = np.atleast_2d(datax_train).T
        datay_train = np.atleast_2d(datay_train).T
        datax_test = np.atleast_2d(datax_test).T
        datay_test = np.atleast_2d(datay_test).T
        
        # Fit the model with training data
        linear_reg.fit(datax_train, datay_train)
        
        # Get the prediction for testing data
        predict = linear_reg.predict(datax_test)
        
        # Get the score
        ll.append((xstr, ystr, r2_score(datay_test, predict)))
```

```python
s = list(set(sorted(l, key=lambda x: x[2], reverse=True)))
s2 = (sorted(ll, key=lambda x: x[2], reverse=True))
```

```python
for t in s:
    if t[2] > 0.65:
        if t[0] == 'bug' or t[1] == 'bug':
                continue
        else:
            print (t[0], t[1], 'Score: ', t[2])
```

```python
for t in s2:
    if t[2] > 0.65:
        if t[1] == 'bug':
                continue
        else:
            print (t[0], t[1], 'Score: ', t[2])
```
